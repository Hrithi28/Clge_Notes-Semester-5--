DataFrame:
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
import pandas as pd
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)
print(df)
print("-----")

#Locate a specific row - 'Name' is index of the row
print(df.loc[2])
print("-----")

#Return a set of rows
print(df.loc[[0,1]])
print("-----")

#To mention specific index name
df = pd.DataFrame(data, index=["d1","d2","d3"])
print(df)
print("-----")
print(df.loc["d1"])
```

```python
#To print the data in a specific column
cal_detail=data["calories"]
print(cal_detail)
```

```python
print(df)
print("-----")
df.describe().T
#displays 5-point summary
df.describe()
```

Distribution Plot Using Seaborn: Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.

```python
import seaborn as sb
#Samuel Norman Seaborn ( sns )
#import seaborn as sns
```

```python
current_palette = sb.color_palette()
sb.palplot(current_palette)
plt.show()
sb.palplot(sb.color_palette("Greens"))
plt.show()
sb.palplot(sb.color_palette("BrBG", 7))
plt.show()
```

```python
#univariant set of observations and visualizes it through a histogram
sb.distplot(df2['Maxpulses'])
```

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sb.histplot(df2['Maxpulses'],kde=True)
```

```python
sb.distplot(df2['Duration'],hist=False)
```

```python
sb.distplot(df2['Calories'],hist=False)
```

```python
#Bivariant - Scatter plot
sb.jointplot(x = 'Calories',y = 'Duration',data = df2)
plt.show()
```

```python
#Bivariant - Scatter plot
#Hexagonal binning is used in bivariate data analysis when the data is sparse in density i.e.,
#when the data is very scattered and difficult to analyze through scatterplots
sb.jointplot(x = 'Calories',y = 'Duration',data = df2, kind='hex')
plt.show()
```

```python
#pairplots - To plot multiple pairwise bivariate distributions in a dataset
#shows the relationship for (n,2) combination of variable in a DataFrame as a matrix of plots and
#the diagonal plots are the univariate plots
sb.set_style("ticks")
sb.pairplot(df,hue="Calories",diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```

Skewness of data: Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean.


*  Positive value: Positively skewed -> Right skewed
*  Negative value: Negatively skewed -> Left skewed
*  Zero value: No skewness





```python
print(df2["Calories"].skew())
print(df2["Duration"].skew())
print(df2["Pulse"].skew())
print(df2["Maxpulses"].skew())
```

```python
plt.figure().set_figheight(20)
#plt.figure().set_figwidth(20)
df2['Duration'].value_counts().plot(kind='pie')
plt.plot()
```

```python
df3 = pd.read_csv('Hotel Reservations.csv')
```

```python
#Categorical Scatter Plot
#Stripplot - when one of the variable under study is categorical
sb.stripplot(x = "type_of_meal_plan", y = "market_segment_type", data = df3)
plt.show()
sb.stripplot(x = "booking_status", y = "room_type_reserved", data = df3)
plt.show()
sb.stripplot(x = "type_of_meal_plan", y = "no_of_children", data = df3, jitter=True) #alternate is swarmplot
plt.show()
```

Box Plot: Outlier identification -visualize the distribution of data through their quartiles

*   Inter Quartile Range(IQR) = Q3-Q1
*   Lower Limit = Q1-1.5 IQR
*   Upper Limit = Q3+1.5 IQR








```python
plt.figure().set_figheight(15)
plt.figure().set_figwidth(20)
sb.boxplot(x=df2['Duration'],y=df2['Calories'])
plt.show
#sb.boxplot(x=df2['Duration'],y=df2['Calories'],orient="h")
#plt.show
```

```python
#Violin Plots are a combination of the box plot with the kernel density estimates
#the wider portion of violin indicates the higher density and narrow region represents relatively lower density
sb.violinplot(x = "no_of_weekend_nights", y = "no_of_week_nights", data = df3)
plt.show()
```

```python
sb.violinplot(x = "no_of_weekend_nights", y = "no_of_week_nights",hue="no_of_special_requests", data = df3)
plt.show()
```

```python
#Bar plot - Bar plot represents the estimate of central tendency
df4 = pd.read_csv('Wholesale customers data.csv')
sb.barplot(x = "Channel", y = "Milk", hue = "Region", data = df4)
plt.show()
```

```python
sb.countplot(x = "Channel",data = df4)
plt.show()
```

```python
sb.pointplot(x = "Channel", y = "Milk", hue = "Region", data = df4)
plt.show()
```
